Task #1 is to first implement gradient descent on a given function $$ y = x^{2} $$ with $$x0 = 4$$, learning rate $$ \eta = 0.1 $$ and tolerance $$ \delta = 0.000001 $$ and converge to the global minimum of the function which is also the only one local minimum since it is a convex function.

Task #2 is to generalize the gradient descent function and define it so that gradient descent can be applied to any function.

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Task #1

```python
f = lambda x: x**2
xpoints = np.linspace(-5,5,100)
ypoints = np.array([f(x) for x in xpoints])
x0 = 4

plt.title("y = x**2")
plt.xlabel("x values")
plt.ylabel("f(x) values")
plt.plot(xpoints,ypoints)
plt.scatter(x0,f(x0), label=x0)

plt.legend()
plt.show()
```

```python
f_ = lambda x: 2*x

tolerance = 0.000001
lr = 0.1
values = [x0,]

for x in range(100):
    xk = values[x] - lr * f_(values[x])
    values.append(xk)
    plt.scatter(xk, f(xk))
    if np.abs(xk - values[x]) < tolerance:
        break
    
plt.title("y = x**2")
plt.xlabel("x values")
plt.ylabel("f(x) values")
plt.plot(xpoints,ypoints)
plt.show()
```

### Task #2

```python
f0 = "x**2"
def gradient_descent(f,f_,x0,lr,tolerance,interval,iterations):
    xpoints = np.linspace(interval[0],interval[1],100)
    ypoints = np.array([f(x) for x in xpoints])
    plt.title(f0)
    plt.xlabel("x values")
    plt.ylabel("f(x) values")
    plt.plot(xpoints,ypoints)
    plt.scatter(x0,f(x0))
    
    values = [x0,]

    for x in range(iterations):
        xk = values[x] - lr * f_(values[x])
        values.append(xk)
        plt.scatter(xk, f(xk))
        if np.abs(xk - values[x]) < tolerance:
            print("Converged to Minimum")
            print("x* = ", xk)
            print("f(x*) = ", f(xk))
            break
    
    plt.show()
```

```python
gradient_descent(lambda x: x**2,lambda x: 2*x,4,0.1,1e-5,(-5,5),100)
```

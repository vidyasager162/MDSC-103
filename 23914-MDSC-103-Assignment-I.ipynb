Solving Optimization Problems from Book using the SciPy Library

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
from mpl_toolkits import mplot3d
from scipy.optimize import minimize
```

```python
f = lambda x : 5*x[0]**2 + x[1]**2 + 2*x[0]*x[1]
cons = ({'type': 'eq', 'fun': lambda x: x[0]*x[1] - 10})

x1 = np.linspace(-10,10,100)
x2 = np.linspace(-10,10,100)

X,Y = np.meshgrid(x1,x2)
F = f((X,Y))

fig = plt.figure(figsize=(8,8))
ax = plt.axes(projection='3d')

ax.set_xlabel('x1', labelpad=10)
ax.set_ylabel('x2', labelpad=30)
ax.set_zlabel('f(X)', labelpad=10)

surf = ax.plot_surface(X,Y,F, cmap='gist_rainbow')
plt.show()
```

```python
x0 = (1,1)
res = minimize(f,x0,constraints=cons)
res
```

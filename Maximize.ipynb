```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize
```

```python
f = lambda x: -5*x[0] - 4*x[1]
cons = ({'type': 'ineq', 'fun': lambda x: -6*x[0] - 4*x[1] + 24},
        {'type': 'ineq', 'fun': lambda x: -x[0] - 2*x[1] + 6},
        {'type': 'ineq', 'fun': lambda x: x[0] - x[1] + 1},
        {'type': 'ineq', 'fun': lambda x: -x[1] + 2})
x0 = (1,1)
res = minimize(f, x0, constraints=cons)
res
```

```python
f = lambda x: 5*x[0] + 4*x[1]
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
X,Y = np.meshgrid(x,y)
F = f((X,Y))

fig = plt.figure(figsize=(12,8))
ax = plt.subplot(projection='3d')
surf = ax.plot_surface(X,Y,F, cmap='jet')
fig.colorbar(surf,shrink=0.4,aspect=10)

ax.set_xlabel('x', labelpad=10)
ax.set_ylabel('y', labelpad=30)
ax.set_zlabel('z', labelpad=10)

plt.show()
```

```python
x = np.linspace(0,5,100)
y = x.copy()

y1 = (24 - 6*x)/4
y2 = (6 - x)/2 
y3 = 1 + x
y4 = 2 - 0*x
y5 = 0*x + 0
x1 = 0*y + 0

plt.figure(figsize=(10,8))
plt.plot(x,y1, label="6x1 + 4x2 <= 24")
plt.plot(x,y2, label="x1 + 2x2 <= 6")
plt.plot(x,y3, label="x2 - x1 <= 1")
plt.plot(x,y4, label="x2 <= 2")
plt.plot(x,y5, label="y >= 0")
plt.plot(x1,y, label="x >= 0")
plt.grid()
plt.ylim(0,5)
plt.legend()

plt.show()
```
